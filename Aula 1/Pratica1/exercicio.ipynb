```python
df['Vendas_Por_Produto'] = df['Quantidade'] * df['Preco_Unitario']
df
```

```python
total_vendas = df.groupby('Produto').sum()['Vendas_Por_Produto']
total_vendas
#Salvar em uma planilha excel 
total_vendas.to_excel('total_vendas.xlsx')
```

```python
#Filtre as vendas do mês de janeiro (1) de 2023 e crie um DataFrame separado com esses dados.
df['Data'] = pd.to_datetime(df['Data'])
df['Ano'] = df['Data'].dt.year
df['Mes'] = df['Data'].dt.month
df_janeiro_2023 = df[(df['Ano'] == 2023) & (df['Mes'] == 1)]
df2 = pd.DataFrame(df_janeiro_2023)
df2
```

```python
df2.to_csv('total_vendas_janeiro.csv')
```
